```python
#Lets see how to grab all the text from a pdf

#step 1: create an empty list
#step 2: create a pdf reader
#step 3: loop through the pdf reader and get each page and append the extracted
#text from each page to the empty list
#step 4: close the file (original pdf from which you are copying)
```

```python
f = open('F:\\NLP intro\\course stuff\\5.1 UPDATED_NLP_COURSE.zip\\UPDATED_NLP_COURSE\\00-Python-Text-Basics\\US_Declaration.pdf', mode ='rb')
```

```python
#lets say we wanted to list all the texts.. so create an empty list first

pdf_text = []

pdf_reader = PyPDF2.PdfReader(f)

for p in range(len(pdf_reader.pages)):
    
    page = pdf_reader.pages[p]
    
    pdf_text.append(page.extract_text())
    
f.close()
```

```python
len(pdf_text)
```

```python
for page in pdf_text:
    print(page)
    print('\n')
    print('\n')
    print('\n')
    print('\n')
    print('\n')
    print('\n')
    print('\n')
    print('\n')
```
